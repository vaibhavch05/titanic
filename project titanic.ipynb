```python

titanic_data = pd.read_csv('titanic.csv')

male_count = titanic_data['sex'].value_counts()['male']
female_count = titanic_data['sex'].value_counts()['female']


labels = ['Males', 'Females']
sizes = [male_count, female_count]
colors = ['#007bff', '#ff69b4']
explode = (0.1, 0)  

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=140, colors=colors)
ax1.axis('equal')  

plt.title('Male vs Female Distribution in Titanic Dataset')
plt.show()
```

```python

titanic_data = pd.read_csv('titanic.csv')


ticket_types = titanic_data['ticket'].apply(lambda x: x.split()[0]).value_counts()


plt.figure(figsize=(10, 6))
ticket_types.plot(kind='bar')
plt.title('Distribution of tickett Types in Titanic Dataset')
plt.xlabel('ticket Type')
plt.ylabel('Count')
plt.show()
```

```python
ticket_types_grouped = titanic_data.groupby('ticket').size().reset_index(name='Count')
print(ticket_types_grouped)
```

```python

titanic_data = pd.read_csv('titanic.csv')

embarked_counts = titanic_data['embarked'].value_counts()


print("Most and Least Populated Home Destinations:")
print(embarked_counts)
```

```python
print("Top 2 Most Populated Home Destinations:")
print(embarked_counts.nlargest(2))

print("\nBottom 2 Least Populated Home Destinations:")
print(embarked_counts.nsmallest(2))
```

```python

titanic_data = pd.read_csv('titanic.csv')

total_parents_children = titanic_data['parch'].sum()

print("Total Number of Parents and Children Aboard:", total_parents_children)
```

```python
import pandas as pd
titanic_data = pd.read_csv('titanic.csv')

embarked_dummies = pd.get_dummies(titanic_data['Embarked'])
correlations = titanic_data['Survived'].corr(embarked_dummies['C']), titanic_data['Survived'].corr(embarked_dummies['Q']), titanic_data['Survived'].corr(embarked_dummies['S'])
print("Correlation of Survived with Embarked:")
print("C:", correlations[0])
print("Q:", correlations[1])
print("S:", correlations[2])

pclass_dummies = pd.get_dummies(titanic_data['pclass'])
correlations = embarked_dummies['C'].corr(pclass_dummies['1']), embarked_dummies['C'].corr(pclass_dummies['2']), embarked_dummies['C'].corr(pclass_dummies['3']), \
               embarked_dummies['Q'].corr(pclass_dummies['1']), embarked_dummies['Q'].corr(pclass_dummies['2']), embarked_dummies['Q'].corr(pclass_dummies['3']), \
               embarked_dummies['S'].corr(pclass_dummies['1']), embarked_dummies['S'].corr(pclass_dummies['2']), embarked_dummies['S'].corr(pclass_dummies['3'])
print("\nCorrelation of embarked with pclass:")
print("C-Pclass1:", correlations[0])
print("C-Pclass2:", correlations[1])
print("C-Pclass3:", correlations[2])
print("Q-Pclass1:", correlations[3])
print("Q-Pclass2:", correlations[4])
print("Q-Pclass3:", correlations[5])
print("S-Pclass1:", correlations[6])
print("S-Pclass2:", correlations[7])
print("S-Pclass3:", correlations[8])


sex_dummies = pd.get_dummies(titanic_data['sex'])
correlations = embarked_dummies['C'].corr(sex_dummies['female']), embarked_dummies['C'].corr(sex_dummies['male']), \
               embarked_dummies['Q'].corr(sex_dummies['female']), embarked_dummies['Q'].corr(sex_dummies['male']), \
               embarked_dummies['S'].corr(sex_dummies['female']), embarked_dummies['S'].corr(sex_dummies['male'])
print("\nCorrelation of embarked with sex:")
print("C-female:", correlations[0])
print("C-male:", correlations[1])
print("Q-female:", correlations[2])
print("Q-male:", correlations[3])
print("S-female:", correlations[4])
print("S-male:", correlations[5])
```
